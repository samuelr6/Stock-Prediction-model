```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
stocks = pd.read_csv('2 Stock_List.csv')
stock_list = stocks['Symbol'].tolist()
print(stocks)
```

```python
test_stocks = pd.DataFrame(np.array([['A.csv', 1, 'A'], ['AA.csv', 2, 'AA']]), columns=['Filename', 'Number', 'Symbol'])
previous_data = pd.read_csv(test_stocks.iloc[0]['Filename'])
new_data = yf.download(test_stocks.iloc[0]['Symbol'], start="2020-04-02", end="2020-12-08")
new_data.reset_index(level='Date', inplace=True)
new_data['Date'] = new_data['Date'].dt.strftime('%m/%d/%Y')
updated_data = previous_data.append(new_data, ignore_index=True)
print(previous_data)
print(new_data)
print(updated_data)
```

```python
test_stocks = pd.DataFrame(np.array([['A.csv', 1, 'A'], ['AA.csv', 2, 'AA']]), columns=['Filename', 'Number', 'Symbol'])
for index, row in test_stocks.iterrows():
    previous_data = pd.read_csv(row['Filename'])
    new_data = yf.download(row['Symbol'], start="2020-04-02", end="2020-12-08")
    new_data.reset_index(level='Date', inplace=True)
    new_data['Date'] = new_data['Date'].dt.strftime('%m/%d/%Y')
    updated_data = previous_data.append(new_data, ignore_index=True)
    updated_data.to_csv(row['Filename'], index=False)
```
